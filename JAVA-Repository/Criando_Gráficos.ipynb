```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Netflix Dataset Latest 2021.xlsx')
```

```python
df = df.head() #Retorna os 5 primeiros valores do df.
df
```

```python
df['Genre'].value_counts() #Contabiliza os valores.
```

```python
eixo_x = df['Genre'].value_counts().keys()
eixo_x
```

```python
eixo_y = df['Genre'].value_counts().values
eixo_y
```

```python
plt.bar(eixo_x, eixo_y)
plt.title("Filmes por gênero")

plt.xticks(rotation = 45)
plt.yticks(rotation = 45)

plt.xlabel('Generos de filmes')
plt.ylabel('Qntd. de filmes')

plt.show()
```

```python
eixo_x = df['Netflix Release Date'].value_counts().sort_index().keys()
eixo_y = df['Netflix Release Date'].value_counts().sort_index().values
```

```python
plt.title('Melhores filmes por ano de lançamento')

plt.xlabel('Qntd. de filmes')
plt.ylabel('Data de lançamento')

plt.plot(eixo_x, eixo_y)
```
